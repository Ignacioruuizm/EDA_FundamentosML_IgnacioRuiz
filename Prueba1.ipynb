```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv('marketing_campaign.csv', delimiter='\t')
print(df.head())
```

```python
df['Edad'] = 2025 - df['Year_Birth']
# Convierte a formato datetime
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y') #convierte dt customer de date a fecha y hora
```

```python
df['Días_Activos'] = (pd.to_datetime('today') - df['Dt_Customer']).dt.days
# Calcula los días desde la inscripción hasta hoy
```

```python
print(df[['Edad', 'Ingreso', 'Días_Activos']].describe())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Edad'], kde=True, bins=20)
plt.title('Distribución de Edad')
plt.show()
# Análisis de distribución
```

DISTRIBUCION DE EDAD:

La distribucion es asimetrica para la derecha, lo que quiere decir que hay mas personas en el rango de edad media que de la avanzada, la mayoria se concentra entre los 40 y 70.

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Ingreso'], kde=True, bins=20)
plt.title('Distribución de Ingreso')
plt.show()
```

DISTRIBUCIÓN DE INGRESO

Se puede ver una asimetria positiva con una concentracion de valores entre 0 y 100000

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Días_Activos'], kde=True, bins=20)
plt.title('Distribución de Días Activos')
plt.show()
```

DISTRIBUCIÓN DE DIAS ACTIVOS:

La distribución de dias activos es mas parecida lo que podria decir que la mayoria de clientes han estado activos por periodos similares.

```python
plt.figure(figsize=(6, 5))
sns.boxplot(y=df['Ingreso'])
plt.title('Boxplot de Ingresos')
plt.show()

plt.figure(figsize=(6, 5))
sns.boxplot(y=df['Edad'])
plt.title('Boxplot de Edad')
plt.show()
```

```python
sns.scatterplot(x='Edad', y='Ingreso', data=df)
plt.title('Relación entre Edad e Ingreso')
plt.xlabel('Edad')
plt.ylabel('Ingreso')
plt.show()
```

RELACION ENTRE EDAD Y INGRESO:
En la relacion entre edad e ingreso se puede observar que entre los 20 y 80 años sus ingresos son menores 200000
